```python
#Density-Based Spatial Clustering of Applications with Noise (DBSCAN)
"""
epsilon: eps. birbirine komşu noktalara bakarken belirlediğimiz mesafe.
min_points: kullanıcı tarafından belirlenen minimum sayıda nokta.
core points: yarıçapın içinde kalanlar
border points: yarıçapın dışında kalanlar
noise: sınırların dışında kalanlar 
"""
```

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('data/dbscan.csv',delimiter = ',')
df = dataset
df.head()
```

```python
plt.figure(figsize = (5,5))
plt.scatter(df["0"], df["1"], s=15, color = 'red')
plt.title('Dataset', fontsize = 20)
plt.xlabel('Feature 1', fontsize = 14)
plt.ylabel('Feature 2', fontsize = 14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN()
dbscan.fit(df[['0', '1']])
```

```python
df['DBSCAN_labels'] = dbscan.labels_
plt.figure(figsize = (5,5))
plt.scatter(df["0"], df["1"], c = df['DBSCAN_labels'], cmap = 'rainbow')
plt.show()
```

```python
from sklearn.neighbors import NearestNeighbors
neigh = NearestNeighbors(n_neighbors = 2)
nbrs = neigh.fit(df[['0', '1']])
distances, indices = nbrs.kneighbors(df[['0', '1']])
print(distances)
print(indices)
```

```python
distances = np.sort(distances, axis = 0)
plt.figure(figsize = (5,5))
plt.plot(distances)
plt.title('K-distance Grafiği', fontsize = 20)
plt.xlabel('Mesafelere göre sıralanmış veri noktaları', fontsize = 14)
plt.ylabel('Epsilon', fontsize = 14)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan_opt = DBSCAN(eps = 4, min_samples = 6)
dbscan_opt.fit(df[['0', '1']])
```

```python
df['DBSCAN_opt_labels'] = dbscan_opt.labels_
plt.figure(figsize = (5,5))
plt.scatter(df["0"], df["1"], c = df['DBSCAN_opt_labels'], cmap = 'rainbow')
plt.show()
```
