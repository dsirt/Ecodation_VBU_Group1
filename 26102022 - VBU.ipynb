```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
fish = pd.read_csv('data/Fish.csv')
fish.head()
```

```python
#PairPlot
plt.figure(figsize = (11,9))
sns.pairplot(fish, hue = 'Species')
plt.show()
```

```python
fish1 = fish[fish['Species'].isin(['Bream', 'Perch', 'Pike'])]
```

```python
plt.figure(figsize = (11,9))
sns.pairplot(fish1, hue = 'Species')
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', palette = 'husl', height = 2)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Length1", "Weight", "Height", "Width"], 
            markers = ['o', 'D', '<'], height = 3, aspect = 1)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], 
            markers = ['o', 'D', '<'], height = 5, aspect = 1)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], kind = 'reg', height = 5, aspect = 1)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], plot_kws = dict(s = 140), height = 5, aspect = 1)
plt.show()
```

```python
sns.pairplot(fish1, hue = 'Species', x_vars = ["Weight", "Height"], y_vars = ["Weight", "Height"],
             plot_kws = dict(s = 140, linewidth = 1, alpha = 0.6), height = 5, aspect = 1)
plt.show()
```

```python
#PairGrid
g = sns.PairGrid(fish1)
g = g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(fish1)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species', palette = "husl")
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species')
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype = "step", linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
g = sns.PairGrid(fish1, hue = 'Species', vars = ["Weight", "Height", "Width"], height = 5, aspect = 1)
g = g.map_offdiag(plt.scatter, edgecolor = "w", s = 130)
g = g.map_diag(plt.hist, edgecolor = "w", linewidth = 2)
g = g.add_legend()
plt.show()
```

```python
insurance = pd.read_csv('data/insurance.csv')
insurance.head()
```

```python
sns.lmplot(x = 'bmi', y = 'expenses', hue = 'smoker', data = insurance, height = 8, aspect = 1.2)
plt.show()
```

```python
sns.lmplot(x = 'bmi', y = 'expenses', hue = 'smoker', data = insurance, height = 8, aspect = 1.2,
          scatter_kws = {'s' : 90, 'alpha' : .4}, palette = ["#FF3333", "#00CC00"])
plt.show()
```

```python
sns.lmplot(x = 'bmi', y = 'expenses', hue = 'smoker', col = 'sex', data = insurance, height = 6, aspect = 1)
plt.show()
```

```python
sns.lmplot(x = 'bmi', y = 'expenses', hue = 'smoker', col = 'region', col_wrap = 2, data = insurance, height = 7, aspect = 1)
plt.show()
```

```python
canada = pd.read_csv('data/canada1.csv', delimiter=';')
canada.head()
```

```python
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1)
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
canada.head()
```

```python
canada1 = canada.loc['2004': , ['Germany', 'Brazil', 'Israel', 'Jamaica', 'Afghanistan', 'Zambia', 'Argentina', 'Austria', 'Uruguay', 'Zimbabwe']]
canada1.head()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 2000)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, center = 4000)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'coolwarm')
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, linewidths = .1)
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, annot = True, fmt = "d")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'YlGnBu', annot = True, fmt = ".2f")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'YlGnBu', linewidths = .5, annot = True, fmt = ".1f")
plt.yticks(rotation = 0)
plt.show()
```

```python
plt.figure(figsize = (16, 9))
ax = sns.heatmap(canada1, vmin = 0, vmax = 8000, cmap = 'YlGnBu', linewidths = .5, annot = True, fmt = ".1f", cbar = False)
plt.yticks(rotation = 0)
plt.show()
```
