```python
def square(n):
    n = n*n
    return n

square(10)
```

```python
def even_odd(num):
    """This function will check whether a number is even or odd""" #docstring
    if num % 2 == 0:
        print(num, ' is an even number')
    else:
        print(num, ' is an odd number')
        
even_odd(3)
even_odd(4)
print(even_odd.__doc__)
```

```python
def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial(num - 1)

factorial(4)
```

```python
def printinfo(name, age = 23):
    print('Name :-', name)
    print('Age :-', age)
    
printinfo(age = 19, name = 'Tasmanian Devil')
printinfo(name = 'Bugs Bunny')
```
