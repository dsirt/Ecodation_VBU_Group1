```python
x = np.arange(1, 31)
y = np.random.normal(10, 11, size = 30)
y = np.square(y)
plt.figure(figsize = (16, 6))
plt.plot(x, y)
plt.fill_between(x, y)
plt.show()
```

```python
x = np.arange(1, 31)
y = np.random.normal(10, 11, size = 30)
y = np.square(y)
plt.figure(figsize = (16, 6))
plt.plot(x, y, color = '#baf1a1')
plt.fill_between(x, y, color = '#7fcd91')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
```

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
df = pd.DataFrame({"C1" : col1, "C2" : col2})
df.head(10)
```

```python
#Lineplot 
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = df.C1, y = df.C2, data = df)
plt.show()
```

```python
plt.figure(figsize = (14, 6))
sns.set(rc = {"axes.facecolor" : "#FFFDE7", "axes.grid" : True, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, linewidth = 2)
sns.lineplot(x = df.C1, y = df.C2 + 1, data = df, linewidth = 2)
plt.show()
```

```python
plt.figure(figsize = (14, 6))
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
sns.lineplot(x = df.C1, y = df.C2, data = df, color = "#FF5722", linewidth = 2)
sns.lineplot(x = df.C1, y = np.cos(df.C1), data = df, color = "#FFEB3B", linewidth = 2)
plt.show()
```

```python
canada = pd.read_csv('data/canada1.csv', delimiter=';')
canada.head(10)
```

```python
canada.columns
```

```python
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1)
```

```python
canada.rename(columns = {'OdName' : 'Country'}, inplace = True)
canada.set_index(canada.Country, inplace = True)
canada.drop(columns = ['Type', 'Coverage', 'AREA', 'AreaName', 'REG', 'RegName', 'DEV', 'DevName'], axis = 1, inplace = True)
canada.head()
```

```python
canada.index.name = None
canada.head()
```

```python
del canada['Country']
canada.head()
```

```python
canada = canada.transpose()
canada.head()
```

```python
plt.figure(figsize = (16,6))
plt.title("Immigrants from Afghanistan", fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 10, 'ytick.labelsize' : 10})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['Afghanistan'])
plt.show()
```

```python
mpl.rcParams.update(mpl.rcParamsDefault)
%matplotlib inline
```

```python
plt.figure(figsize = (16,6))
plt.title("Immigrants from India, Pakistan, Bangladesh", fontsize = 14)
sns.set(rc = {"axes.facecolor" : "#283747", "axes.grid" : False, 'xtick.labelsize' : 14, 'ytick.labelsize' : 14})
plt.xticks(rotation = 45)
sns.lineplot(x = canada.index.values, y = canada['India'], color = '#ff9900', label = 'India')
sns.lineplot(x = canada.index.values, y = canada['Pakistan'], color = '#4586ff', label = 'Pakistan')
sns.lineplot(x = canada.index.values, y = canada['Bangladesh'], color = '#a2ef44', label = 'Bangladesh')
plt.legend(facecolor = 'grey', fontsize = 'large', edgecolor = 'black', shadow = True)
plt.show()
```

```python
employment = pd.read_excel('data/employmentx2005v2.xlsx')
employment.head()
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = "Period", y = "Unemployed", hue = "Gender", data = employment)
plt.show()
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = "Period", y = "Unemployed", hue = "Gender", style = "Gender", markers = True, data = employment)
plt.show()
```

```python
plt.figure(figsize = (14,7))
plt.style.use('seaborn-darkgrid')
sns.lineplot(x = "Period", y = "Unemployed", hue = "Age", style = "Gender", estimator = None, units = "Age", markers = True, data = employment)
plt.legend(bbox_to_anchor = (1.0, 1.0), shadow = True, fontsize = 'large')
plt.show()
```
