```python
daf1 = pd.DataFrame({'id' : ['1', '2', '3', '4', '5'], 'Name' : ['Doğu', 'Batı', 'Kuzey', 'Güney', 'Güneş']})
daf1
```

```python
daf2 = pd.DataFrame({'id' : ['1', '2', '6', '7', '8'], 'Score' : ['40', '50', '80', '90', '70']})
daf2
```

```python
pd.merge(daf1, daf2, on = 'id', how = 'inner') #INNER JOIN
```

```python
pd.merge(daf1, daf2, on = 'id', how = 'outer') #OUTER JOIN
```

```python
pd.merge(daf1, daf2, on = 'id', how = 'left')
```

```python
pd.merge(daf1, daf2, on = 'id', how = 'right')
```
